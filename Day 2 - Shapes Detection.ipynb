```python
import cv2

def find_triangles(image_path):
    img = cv2.imread(image_path)
    
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    
    blurred = cv2.GaussianBlur(gray, (7, 7), 0)
    
    edges = cv2.Canny(blurred, 30, 150)

    contours, _ = cv2.findContours(edges, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    
    image_with_boxes=img.copy()
    
    for contour in contours:
        epsilon = 0.04 * cv2.arcLength(contour, True)
        approx = cv2.approxPolyDP(contour, epsilon, True)
    
        if len(approx) == 3:
            x, y, w, h= cv2.boundingRect(contour)
            cv2.rectangle(image_with_boxes,(x,y),(x+w,y+h),(0,0,255),2)
        
    cv2.imshow('Triangles with Bounding Boxes', image_with_boxes)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
    
find_triangles(r'C:\Users\sasra\Downloads\CS Study\DBA\img.jpg')
```
